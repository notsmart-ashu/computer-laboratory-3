```python
import random

# Fitness Function
# Objective: Maximize f(x) = x^2

def fitness(x):
    return x * x

# Input parameters
population_size = int(input("Enter population size: "))
generations = int(input("Enter number of generations: "))
clone_factor = int(input("Enter clone factor: "))
mutation_rate = float(input("Enter mutation rate (0 to 1): "))

# Generate initial population
population = []

for i in range(population_size):
    antibody = random.randint(1, 20)
    population.append(antibody)

print("\nInitial Population:")
print(population)

# Clonal Selection Process
for generation in range(generations):

    print(f"\nGeneration {generation + 1}")

    # Evaluate fitness
    fitness_values = []

    for antibody in population:
        fit = fitness(antibody)
        fitness_values.append(fit)

    # Select best antibody
    best_index = fitness_values.index(max(fitness_values))
    best_antibody = population[best_index]

    print("Best Antibody =", best_antibody)
    print("Best Fitness =", fitness(best_antibody))

    # Cloning
    clones = []

    for i in range(clone_factor):
        clones.append(best_antibody)

    print("Clones =", clones)

    # Mutation
    mutated_clones = []

    for clone in clones:

        if random.random() < mutation_rate:

            mutation = random.randint(-2, 2)

            clone = clone + mutation

            if clone < 1:
                clone = 1

        mutated_clones.append(clone)

    print("Mutated Clones =", mutated_clones)

    # Replace old population
    population = mutated_clones

# Final Result
best_solution = max(population, key=fitness)

print("\nFinal Best Solution =", best_solution)
print("Final Best Fitness =", fitness(best_solution))
```
