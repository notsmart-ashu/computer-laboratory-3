```python
import random
import numpy as np

# Sample dataset for spray drying process
# Input: Temperature, Feed Flow Rate
# Output: Powder Yield

X = np.array([
    [150, 20],
    [160, 25],
    [170, 30],
    [180, 35],
    [190, 40]
])

Y = np.array([60, 65, 72, 80, 90])

# Genetic Algorithm Parameters
population_size = int(input("Enter population size: "))
mutation_rate = float(input("Enter mutation rate (0 to 1): "))
generations = int(input("Enter number of generations: "))

# Create initial population
population = []

for i in range(population_size):

    # Random weights
    individual = np.random.rand(2)

    population.append(individual)

# Fitness Function
def fitness(individual):

    predictions = []

    for x in X:

        prediction = np.dot(x, individual)

        predictions.append(prediction)

    predictions = np.array(predictions)

    error = np.mean((Y - predictions) ** 2)

    return 1 / (error + 1)

# Genetic Algorithm Process
for generation in range(generations):

    fitness_values = []

    for individual in population:

        fit = fitness(individual)

        fitness_values.append(fit)

    # Select best individual
    best_index = np.argmax(fitness_values)

    best_individual = population[best_index]

    print(f"\nGeneration {generation + 1}")
    print("Best Fitness =", fitness_values[best_index])
    print("Best Parameters =", best_individual)

    # Create new population
    new_population = []

    for i in range(population_size):

        child = best_individual.copy()

        # Mutation
        if random.random() < mutation_rate:

            mutation = np.random.normal(0, 0.1, size=2)

            child += mutation

        new_population.append(child)

    population = new_population

# Final optimized parameters
print("\nOptimized Parameters:")
print(best_individual)
```
