```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Sample dataset
# Features:
# [Crack Length, Vibration Level]

X = np.array([
    [2, 20],
    [3, 25],
    [1, 15],
    [8, 80],
    [9, 90],
    [7, 75],
    [2, 18],
    [8, 85]
])

# Labels
# 0 = Undamaged
# 1 = Damaged

Y = np.array([0, 0, 0, 1, 1, 1, 0, 1])

# Split dataset
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=42
)

# Artificial Immune Pattern Recognition
# Using similarity-based classification

classifier = KNeighborsClassifier(n_neighbors=3)

# Train model
classifier.fit(X_train, Y_train)

# Predictions
predictions = classifier.predict(X_test)

# Accuracy
accuracy = accuracy_score(Y_test, predictions)

print("Predicted Classes =", predictions)

print("Actual Classes =", Y_test)

print("\nAccuracy =", accuracy * 100, "%")

# User input for testing
print("\nEnter structure details for prediction")

crack = float(input("Enter crack length: "))
vibration = float(input("Enter vibration level: "))

test_sample = np.array([[crack, vibration]])

result = classifier.predict(test_sample)

if result[0] == 0:
    print("\nStructure Status: Undamaged")
else:
    print("\nStructure Status: Damaged")
```
