```python
import random

# Input number of servers
num_servers = int(input("Enter number of servers: "))

servers = []

# Initialize server loads
for i in range(num_servers):
    servers.append(0)

# Input number of client requests
num_requests = int(input("Enter number of client requests: "))

print("\nChoose Load Balancing Algorithm")
print("1. Round Robin")
print("2. Least Connections")
print("3. Random Selection")

choice = int(input("Enter your choice: "))

# ROUND ROBIN
if choice == 1:

    print("\n--- Round Robin Load Balancing ---")

    for request in range(num_requests):

        server_index = request % num_servers

        servers[server_index] += 1

        print(f"Request {request + 1} assigned to Server {server_index + 1}")

# LEAST CONNECTIONS
elif choice == 2:

    print("\n--- Least Connections Load Balancing ---")

    for request in range(num_requests):

        server_index = servers.index(min(servers))

        servers[server_index] += 1

        print(f"Request {request + 1} assigned to Server {server_index + 1}")

# RANDOM SELECTION
elif choice == 3:

    print("\n--- Random Load Balancing ---")

    for request in range(num_requests):

        server_index = random.randint(0, num_servers - 1)

        servers[server_index] += 1

        print(f"Request {request + 1} assigned to Server {server_index + 1}")

else:
    print("Invalid Choice")

# Final Server Load
print("\nFinal Server Loads")

for i in range(num_servers):
    print(f"Server {i + 1} Load = {servers[i]}")
```
