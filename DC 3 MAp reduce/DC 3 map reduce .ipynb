```python
from collections import defaultdict

# Input number of records
n = int(input("Enter number of weather records: "))

weather_data = []

print("\nEnter data in format: Year Temperature")

for i in range(n):

    year = int(input("Enter Year: "))
    temp = float(input("Enter Temperature: "))

    weather_data.append((year, temp))

# MAP PHASE
mapped_data = []

for year, temp in weather_data:

    mapped_data.append((year, temp))

print("\nMapped Data:")
print(mapped_data)

# REDUCE PHASE
temperature_data = defaultdict(list)

for year, temp in mapped_data:

    temperature_data[year].append(temp)

average_temperature = {}

print("\nYearly Average Temperatures:")

for year in temperature_data:

    avg_temp = sum(temperature_data[year]) / len(temperature_data[year])

    average_temperature[year] = avg_temp

    print(f"{year} -> {avg_temp:.2f}")

# Find hottest and coolest year
hottest_year = max(average_temperature, key=average_temperature.get)

coolest_year = min(average_temperature, key=average_temperature.get)

print("\nHottest Year:")
print(
    hottest_year,
    "with Average Temperature =",
    round(average_temperature[hottest_year], 2)
)

print("\nCoolest Year:")
print(
    coolest_year,
    "with Average Temperature =",
    round(average_temperature[coolest_year], 2)
)
```
